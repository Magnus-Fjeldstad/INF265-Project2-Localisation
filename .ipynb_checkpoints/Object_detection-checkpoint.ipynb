# Detection

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from datetime import datetime

torch.manual_seed(123)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Training on device: {device}")

# Image / grid setup
H_in, W_in = 48, 60
H_out, W_out = 2, 3
NUM_CLASSES = 2
OUT_CHANNELS = NUM_CLASSES + 5


class DetectionCNN(nn.Module):
    """
    Fully convolutional detector.
    Output shape: (N, C+5, H_out, W_out)
    """
    def __init__(self, channels=(16, 32, 64), head_channels=128, dropout=0.1, num_classes=NUM_CLASSES):
        super().__init__()

        blocks = []
        in_ch = 1
        for out_ch in channels:
            blocks.extend([
                nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1),
                nn.BatchNorm2d(out_ch),
                nn.ReLU(inplace=True),
                nn.MaxPool2d(2),
            ])
            in_ch = out_ch

        self.backbone = nn.Sequential(*blocks)
        self.grid_pool = nn.AdaptiveAvgPool2d((H_out, W_out))
        self.head = nn.Sequential(
            nn.Conv2d(in_ch, head_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Dropout2d(dropout),
            nn.Conv2d(head_channels, num_classes + 5, kernel_size=1),
        )

    def forward(self, x):
        x = self.backbone(x)
        x = self.grid_pool(x)
        return self.head(x)


model_variants = {
    'small': {'channels': (16, 32), 'head_channels': 64, 'dropout': 0.10},
    'base': {'channels': (16, 32, 64), 'head_channels': 128, 'dropout': 0.10},
    'deep': {'channels': (16, 32, 64, 96), 'head_channels': 160, 'dropout': 0.15},
}


def _zero_if_empty(x, device):
    return x.mean() if x.numel() > 0 else torch.tensor(0.0, device=device)


def detection_loss(y_pred, y_true, lambda_noobj=0.25, lambda_bbox=2.0, lambda_cls=1.0):
    """
    y_pred: (N, C+5, H_out, W_out)
    y_true: (N, H_out, W_out, 6)
    """
    y_pred = y_pred.permute(0, 2, 3, 1)  # -> (N, H_out, W_out, C+5)

    obj_true = y_true[..., 0].float()
    obj_logits = y_pred[..., 0]

    # Objectness
    bce_map = F.binary_cross_entropy_with_logits(obj_logits, obj_true, reduction='none')
    obj_mask = obj_true > 0.5
    noobj_mask = ~obj_mask

    obj_loss = _zero_if_empty(bce_map[obj_mask], y_pred.device)
    noobj_loss = _zero_if_empty(bce_map[noobj_mask], y_pred.device)
    det_loss = obj_loss + lambda_noobj * noobj_loss

    # Bounding boxes (x,y in [0,1], w,h positive and can exceed 1)
    pred_xy = torch.sigmoid(y_pred[..., 1:3])
    pred_wh = F.softplus(y_pred[..., 3:5])
    pred_box = torch.cat([pred_xy, pred_wh], dim=-1)
    true_box = y_true[..., 1:5].float()

    bbox_loss = _zero_if_empty(
        F.smooth_l1_loss(pred_box[obj_mask], true_box[obj_mask], reduction='none').mean(dim=-1),
        y_pred.device,
    )

    # Classes
    class_logits = y_pred[..., 5:]  # (..., NUM_CLASSES)
    class_targets = y_true[..., 5].long().clamp(min=0, max=NUM_CLASSES - 1)
    cls_loss = _zero_if_empty(F.cross_entropy(class_logits[obj_mask], class_targets[obj_mask], reduction='none'), y_pred.device)

    return det_loss + lambda_bbox * bbox_loss + lambda_cls * cls_loss


def _xywh_iou_tensor(box1, box2):
    """
    box1, box2: (..., 4), format [x, y, w, h] in same coordinate frame.
    """
    b1_min = box1[..., :2] - 0.5 * box1[..., 2:]
    b1_max = box1[..., :2] + 0.5 * box1[..., 2:]
    b2_min = box2[..., :2] - 0.5 * box2[..., 2:]
    b2_max = box2[..., :2] + 0.5 * box2[..., 2:]

    inter_min = torch.maximum(b1_min, b2_min)
    inter_max = torch.minimum(b1_max, b2_max)
    inter_wh = (inter_max - inter_min).clamp(min=0)
    inter = inter_wh[..., 0] * inter_wh[..., 1]

    area1 = (box1[..., 2] * box1[..., 3]).clamp(min=0)
    area2 = (box2[..., 2] * box2[..., 3]).clamp(min=0)
    union = area1 + area2 - inter
    return torch.where(union > 0, inter / union, torch.zeros_like(union))


def local_to_global(row, col, bb, cols=W_out, rows=H_out):
    x, y, w, h = bb
    cell_w = 1.0 / cols
    cell_h = 1.0 / rows
    gx = (col + x) * cell_w
    gy = (row + y) * cell_h
    gw = w * cell_w
    gh = h * cell_h
    return gx, gy, gw, gh


def xywh_to_xyxy(bb):
    x, y, w, h = bb
    x1 = x - 0.5 * w
    y1 = y - 0.5 * h
    x2 = x + 0.5 * w
    y2 = y + 0.5 * h
    return x1, y1, x2, y2

```

### Load data and preprocessing

```python
data_path = '../data_2/'

# Load preprocessed detection data (2x3 grid)
train_raw = torch.load(f"{data_path}detection_train.pt")
val_raw = torch.load(f"{data_path}detection_val.pt")
test_raw = torch.load(f"{data_path}detection_test.pt")


def _to_tensors(data_obj):
    """
    Convert loaded object to (images, targets) tensors.
    Supports common formats: TensorDataset, (images, targets), list of tuples.
    """
    if hasattr(data_obj, 'tensors'):
        images, targets = data_obj.tensors
    elif isinstance(data_obj, (tuple, list)) and len(data_obj) == 2 and torch.is_tensor(data_obj[0]) and torch.is_tensor(data_obj[1]):
        images, targets = data_obj
    else:
        pairs = list(data_obj)
        images = torch.stack([p[0] for p in pairs], dim=0)
        targets = torch.stack([p[1] for p in pairs], dim=0)

    images = images.float()
    targets = targets.float()

    if images.ndim == 3:
        images = images.unsqueeze(1)

    return images, targets


train_images, train_targets = _to_tensors(train_raw)
val_images, val_targets = _to_tensors(val_raw)
test_images, test_targets = _to_tensors(test_raw)

# Normalize using train statistics
train_mean = train_images.mean()
train_std = train_images.std().clamp_min(1e-6)

print(f"Train size: {train_images.shape[0]}")
print(f"Val size:   {val_images.shape[0]}")
print(f"Test size:  {test_images.shape[0]}")
print(f"Train mean/std: {train_mean.item():.4f}/{train_std.item():.4f}")


class DetectionDataset(Dataset):
    def __init__(self, images, targets, mean, std):
        self.images = images
        self.targets = targets
        self.mean = mean
        self.std = std

    def __len__(self):
        return self.images.shape[0]

    def __getitem__(self, idx):
        img = (self.images[idx] - self.mean) / self.std
        tgt = self.targets[idx]
        return img, tgt


batch_size = 128
train_loader = DataLoader(DetectionDataset(train_images, train_targets, train_mean, train_std), batch_size=batch_size, shuffle=True)
val_loader = DataLoader(DetectionDataset(val_images, val_targets, train_mean, train_std), batch_size=batch_size, shuffle=False)
test_loader = DataLoader(DetectionDataset(test_images, test_targets, train_mean, train_std), batch_size=batch_size, shuffle=False)

```

### Training

```python
def train_one_model(model, train_loader, val_loader, num_epochs=30, lr=1e-3):
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)

    history = {'train_loss': [], 'val_loss': []}
    best_val_loss = float('inf')
    best_state = None

    for epoch in range(num_epochs):
        model.train()
        train_loss_sum = 0.0

        for images, targets in train_loader:
            images = images.to(device)
            targets = targets.to(device)

            optimizer.zero_grad()
            preds = model(images)
            loss = detection_loss(preds, targets)
            loss.backward()
            optimizer.step()

            train_loss_sum += loss.item() * images.size(0)

        train_loss = train_loss_sum / len(train_loader.dataset)

        model.eval()
        val_loss_sum = 0.0
        with torch.no_grad():
            for images, targets in val_loader:
                images = images.to(device)
                targets = targets.to(device)
                preds = model(images)
                loss = detection_loss(preds, targets)
                val_loss_sum += loss.item() * images.size(0)

        val_loss = val_loss_sum / len(val_loader.dataset)
        scheduler.step()

        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_state = {k: v.detach().cpu().clone() for k, v in model.state_dict().items()}

        print(
            f"Epoch {epoch + 1:02d}/{num_epochs} | "
            f"train: {train_loss:.4f} | val: {val_loss:.4f}"
        )

    if best_state is not None:
        model.load_state_dict(best_state)

    return model, history


num_epochs = 30
trained_models = {}
training_histories = {}

for model_name, cfg in model_variants.items():
    print(f"\nTraining variant: {model_name}")
    net = DetectionCNN(**cfg).to(device)
    net, hist = train_one_model(net, train_loader, val_loader, num_epochs=num_epochs, lr=1e-3)
    trained_models[model_name] = net
    training_histories[model_name] = hist

# Default model reference (best model selected in evaluation cell)
model = trained_models['base'] if 'base' in trained_models else next(iter(trained_models.values()))

```

### Prediction

```python
def _decode_true_boxes(target_grid):
    """
    target_grid: (H_out, W_out, 6)
    Returns list of (global_xywh, label).
    """
    boxes = []
    for r in range(H_out):
        for c in range(W_out):
            cell = target_grid[r, c]
            if cell[0] > 0.5:
                gbox = local_to_global(r, c, cell[1:5])
                label = int(cell[5].item())
                boxes.append((gbox, label))
    return boxes


def _decode_pred_boxes(pred_grid, obj_threshold=0.5):
    """
    pred_grid: (C+5, H_out, W_out)
    Returns list of (global_xywh, label, score).
    """
    pred_grid = pred_grid.permute(1, 2, 0)  # -> (H_out, W_out, C+5)
    boxes = []
    for r in range(H_out):
        for c in range(W_out):
            cell = pred_grid[r, c]
            obj_score = torch.sigmoid(cell[0]).item()
            if obj_score > obj_threshold:
                px = torch.sigmoid(cell[1]).item()
                py = torch.sigmoid(cell[2]).item()
                pw = F.softplus(cell[3]).item()
                ph = F.softplus(cell[4]).item()
                class_probs = F.softmax(cell[5:], dim=0)
                label = int(torch.argmax(class_probs).item())
                score = obj_score * class_probs[label].item()
                gbox = local_to_global(r, c, torch.tensor([px, py, pw, ph]))
                boxes.append((gbox, label, score))
    return boxes


def _draw_box(ax, xywh, color, text=None):
    x, y, w, h = xywh
    x1 = (x - 0.5 * w) * W_in
    y1 = (y - 0.5 * h) * H_in
    rect_w = w * W_in
    rect_h = h * H_in
    rect = patches.Rectangle((x1, y1), rect_w, rect_h, linewidth=2, edgecolor=color, facecolor='none')
    ax.add_patch(rect)
    if text is not None:
        ax.text(x1, max(0, y1 - 2), text, color=color, fontsize=8, backgroundcolor='black')


plot_model = globals().get('best_model', globals().get('model', None))
if plot_model is None:
    raise RuntimeError('No trained model found. Run training/evaluation cells first.')

plot_model.eval()

splits = [
    ('Train', train_images, train_targets),
    ('Validation', val_images, val_targets),
    ('Test', test_images, test_targets),
]

samples_per_split = 3
fig, axes = plt.subplots(len(splits), samples_per_split, figsize=(4.2 * samples_per_split, 3.6 * len(splits)))

with torch.no_grad():
    for row, (split_name, imgs, tgts) in enumerate(splits):
        idxs = torch.randperm(imgs.shape[0])[:samples_per_split]
        for col, idx in enumerate(idxs):
            ax = axes[row, col] if len(splits) > 1 else axes[col]

            img = imgs[idx].float()
            tgt = tgts[idx]

            inp = ((img - train_mean) / train_std).unsqueeze(0).to(device)
            pred = plot_model(inp).cpu().squeeze(0)

            true_boxes = _decode_true_boxes(tgt)
            pred_boxes = _decode_pred_boxes(pred, obj_threshold=0.5)

            ax.imshow(img.squeeze(0), cmap='gray')

            true_labels = []
            pred_labels = []

            for gbox, lab in true_boxes:
                true_labels.append(str(lab))
                _draw_box(ax, gbox, color='lime', text=f"T:{lab}")

            for gbox, lab, score in pred_boxes:
                pred_labels.append(str(lab))
                _draw_box(ax, gbox, color='red', text=f"P:{lab} ({score:.2f})")

            ax.set_title(
                f"{split_name} | true [{', '.join(true_labels) if true_labels else '-'}] "
                f"pred [{', '.join(pred_labels) if pred_labels else '-'}]",
                fontsize=8,
            )
            ax.axis('off')

plt.tight_layout()
plt.show()

```

### Model selection and evaluation

```python
def compute_detection_accuracy(model, loader):
    """
    Accuracy over cells containing an object in y_true.
    A cell is correct if predicted objectness and class are both correct.
    """
    model.eval()
    total_obj_cells = 0
    correct = 0

    with torch.no_grad():
        for images, targets in loader:
            images = images.to(device)
            targets = targets.to(device)
            preds = model(images).permute(0, 2, 3, 1)  # (N, H_out, W_out, C+5)

            true_obj = targets[..., 0] > 0.5
            pred_obj = torch.sigmoid(preds[..., 0]) > 0.5
            pred_cls = preds[..., 5:].argmax(dim=-1)
            true_cls = targets[..., 5].long()

            total_obj_cells += true_obj.sum().item()
            correct += (true_obj & pred_obj & (pred_cls == true_cls)).sum().item()

    return correct / total_obj_cells if total_obj_cells > 0 else 0.0


def compute_detection_iou(model, loader):
    """
    Mean IoU over true object cells. If object is missed, IoU for that cell is 0.
    """
    model.eval()
    total_iou = 0.0
    total_obj_cells = 0

    with torch.no_grad():
        for images, targets in loader:
            images = images.to(device)
            targets = targets.to(device)
            preds = model(images).permute(0, 2, 3, 1)  # (N, H_out, W_out, C+5)

            true_obj = targets[..., 0] > 0.5
            if not true_obj.any():
                continue

            pred_obj = torch.sigmoid(preds[..., 0]) > 0.5

            pred_xy = torch.sigmoid(preds[..., 1:3])
            pred_wh = F.softplus(preds[..., 3:5])
            pred_box = torch.cat([pred_xy, pred_wh], dim=-1)
            true_box = targets[..., 1:5]

            iou_map = _xywh_iou_tensor(pred_box, true_box)
            iou_map = iou_map * pred_obj.float()  # missed detections -> 0 IoU

            total_iou += iou_map[true_obj].sum().item()
            total_obj_cells += true_obj.sum().item()

    return total_iou / total_obj_cells if total_obj_cells > 0 else 0.0


if 'trained_models' not in globals() or len(trained_models) == 0:
    raise RuntimeError('No trained models found. Run the training cell first.')

validation_scores = {}
for model_name, net in trained_models.items():
    val_acc = compute_detection_accuracy(net, val_loader)
    val_iou = compute_detection_iou(net, val_loader)
    val_overall = 0.5 * (val_acc + val_iou)
    validation_scores[model_name] = {
        'accuracy': val_acc,
        'iou': val_iou,
        'overall': val_overall,
    }

print('Validation scores:')
for name, m in validation_scores.items():
    print(f"{name:>8s} | acc: {m['accuracy']:.4f} | iou: {m['iou']:.4f} | overall: {m['overall']:.4f}")

best_model_name = max(validation_scores, key=lambda k: validation_scores[k]['overall'])
best_model = trained_models[best_model_name]
model = best_model  # convenient common alias

print(f"\nSelected best model: {best_model_name}")

# Evaluate best model on test set
test_acc = compute_detection_accuracy(best_model, test_loader)
test_iou = compute_detection_iou(best_model, test_loader)
test_overall = 0.5 * (test_acc + test_iou)

print(f"Test accuracy: {test_acc:.4f}")
print(f"Test IoU:      {test_iou:.4f}")
print(f"Test overall:  {test_overall:.4f}")

```

```python
from torchmetrics.detection.mean_ap import MeanAveragePrecision

def get_map_results(model, eval_loader, device):
    '''
        Helper functions to get predictions and targets in the format required for mAP calculation.
        Depending on your data processing and model architecture this function can either be used as is, 
        modified to fit your needs or used as a blue print for a rewrite.
        Here it is assussmed that the image has been divide into a 2 x 3 grid.
        ----------------------------------------------------------
        Run through the data in the dataloader and collect predicitions and targets for mAP calculation.

        torchmetric mAP expects predictions and targets in the format:
        preds = [
           { "boxes": tensor([[x1, y1, x2, y2], ...]), "scores": tensor([score1, score2, ...]), "labels": tensor([label1, label2, ...])},
            ...   ]
        and targets = [
            { "boxes": tensor([[x1, y1, x2, y2], ...]), "labels": tensor([label1, label2, ...])},
            ...   ]
        where each dict in the list corresponds to one image in the dataset and contains the predicted and true results
    '''
    model.eval()
    with torch.no_grad():
        preds = []
        targets = []
        for images, labels in eval_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            outputs = outputs.permute(0, 2, 3, 1)                               # (B, 7, 2, 3) → (B, 2, 3, 7)
            outputs = outputs.reshape(outputs.shape[0], -1, outputs.shape[-1])  # (B, 2, 3, 7) → (B, 6, 7)
            labels = labels.reshape(labels.shape[0], -1, labels.shape[-1])      # (B, 2, 3, 6) → (B, 6, 6)
            for output, label in zip(outputs, labels):
                pred_boxes = []
                pred_scores = []
                pred_labels = []
                target_boxes = []
                target_labels = []
                # collect predicted boxes, scores and labels for the current image
                for i, cell_output in enumerate(output):
                    pred_object_presence = (torch.sigmoid(cell_output[0]) > 0.5) * 1.0
                    if pred_object_presence == 1:
                        # get propability of object presence and class probabilities to compute detection score for mAP calculation
                        obj_prop = torch.sigmoid(cell_output[0]).item()
                        class_prop = F.softmax(cell_output[5:], dim=0)
                        pred_label = torch.argmax(class_prop)
                        detect_score = obj_prop * class_prop[pred_label]
                        # convert from local to global coordinates before we can compare with the labels and compute IoU for mAP calculation
                        bbox_global = local_to_global(i // 3, i % 3, cell_output[1:5])
                        bbox_xyxy = xywh_to_xyxy(bbox_global)
                        bbox_xyxy = torch.stack(bbox_xyxy)
                        # collect predicted boxes, scores and labels for the current image
                        pred_boxes.append(bbox_xyxy)
                        pred_scores.append(detect_score)
                        pred_labels.append(pred_label)
                # collect true boxes and labels for the current image
                for i, cell_label in enumerate(label):
                    true_object_presence = cell_label[0]
                    if true_object_presence == 1:
                        bbox_global = local_to_global(i // 3, i % 3, cell_label[1:5])
                        bbox_xyxy = xywh_to_xyxy(bbox_global)
                        bbox_xyxy = torch.stack(bbox_xyxy)
                        target_boxes.append(bbox_xyxy)
                        target_labels.append(int(cell_label[-1]))
                # store predictions and targets for the current image in the format required for mAP calculation
                # if there are no predicted boxes, we need to create an empty tensor for the boxes, scores and labels to avoid errors in the mAP calculation
                if len(pred_boxes) == 0:
                    pred_dict = {
                        "boxes": torch.zeros((0, 4), device=device),
                        "scores": torch.zeros((0,), device=device),
                        "labels": torch.zeros((0,), dtype=torch.long, device=device),
                    }
                    preds.append(pred_dict)
                else:
                    pred_dict = {
                        "boxes": torch.stack(pred_boxes),
                        "scores": torch.tensor(pred_scores, device=device),
                        "labels": torch.tensor(pred_labels, device=device),
                    }
                    preds.append(pred_dict)
                # if there are no true boxes, we need to create an empty tensor for the boxes and labels to avoid errors in the mAP calculation            
                if len(target_boxes) == 0:
                    target_dict = {
                        "boxes": torch.zeros((0, 4), device=device),
                        "labels": torch.zeros((0,), dtype=torch.long, device=device),
                    }
                    targets.append(target_dict)
                else:
                    target_dict = {
                        "boxes": torch.stack(target_boxes),
                        "labels": torch.tensor(target_labels, device=device),
                    }
                    targets.append(target_dict)
    
    # compute mAP using torchmetrics
    metric = MeanAveragePrecision(iou_type="bbox")
    metric.update(preds, targets)
    results = metric.compute()
    # results is a dict with the mAP results for different IoU thresholds and the overall mAP
    return results        

def local_to_global(i, j, bb, width=60, height=48, cols=3, rows=2):
    x, y, w, h = bb
    # get the dimensions of a single grid cell
    cell_width, cell_height = width / cols, height / rows
    # convert from local to global coordinates
    global_x = x * cell_width + j * cell_width
    global_y = y * cell_height + i * cell_height
    global_w = w * cell_width
    global_h = h * cell_height

    return global_x, global_y, global_w, global_h

def xywh_to_xyxy(bb):
    # convert from center format to box format
    x_center, y_center, w, h = bb
    x1 = x_center - w/2
    y1 = y_center - h/2
    x2 = x_center + w/2
    y2 = y_center + h/2
    return x1, y1, x2, y2   
```
